```python
import pandas as pd
import numpy as np
from nba_api.stats.endpoints import boxscoretraditionalv2
import random
import time
```

```python
team_df = pd.read_csv('nba_api_data/teamgamelogs.csv')
starting_or_not = pd.read_csv('nba_api_data/lineupdata.csv')
starters_vs_bench = pd.read_csv('nba_api_data/startersvsbench.csv')

game_ids = team_df.loc[~team_df['GAME_ID'].isin(starting_or_not['GAME_ID']), 'GAME_ID'].unique()
np.random.shuffle(game_ids)
i = 0

for index, id in enumerate(game_ids):
    try:
        t = '00' + str(id)
        response = boxscoretraditionalv2.BoxScoreTraditionalV2(game_id=t)
        response = response.get_data_frames()
        temp = response[0]
        temp = temp[['GAME_ID', 'PLAYER_ID', 'START_POSITION']]
        starting_or_not = pd.concat([starting_or_not, temp])
        starters_vs_bench = pd.concat([starters_vs_bench, response[2]])
        i = 0
        print(f'{index + 1}/{len(game_ids)}')
    except Exception as e:
        print(f'Failed to gather response for game: {id}')
        print(e)
        i += 1
        time.sleep(random.uniform(10, 20))
    if i >= 5:
        break

    sleep_time = np.clip(np.random.normal(loc=6, scale=2), 3, 12)
    time.sleep(sleep_time)

    if index % 20 == 0 and index != 0:
        long_sleep = random.uniform(20, 40)
        print(f"Taking a longer break for {long_sleep:.2f} seconds...")
        time.sleep(long_sleep)


starting_or_not.to_csv('nba_api_data/lineupdata.csv')

starters_vs_bench.to_csv('nba_api_data/startersvsbench.csv')
```
